```python
import numpy as np
import pandas as pd
```

# Исходный Data Frame

```python
path = './_data.csv'
df = pd.read_csv(path)
df.head()
```

# Преобразование названий колонок

```python
df.columns = ['Serial_number', 'Announcement ID', 'Number_of_rooms', 'Type', 'Metro', 'Address', 'Area,m2', 'Building_info', 'Parking', 'Price', 'Phone_number', 'Description', 'Interior_finishing',
              'area_of_rooms_m2', 'Balcony', 'Windows', 'Bathroom', 'Children/pets_allowed', 'Additionally', 'Name_of_the_residential_complex', 'Home_series', 'Ceiling_height_m', 'Elevator', 'Garbage_chute', 'Link_to_ad']
```

# Избавляемся от пропусков

```python
#df['Price'] = df['Price'].str.split(' ').str[0]
df['Area,m2'] = df['Area,m2'].str.split('/').str[0]
df['Number_of_rooms'] = df['Number_of_rooms'].str.split(',').str[0]
#df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Area,m2'] = pd.to_numeric(df['Area,m2'], errors='coerce')
df['Number_of_rooms'] = pd.to_numeric(df['Number_of_rooms'], errors='coerce')
df.head()
```

```python
# Заолним пропуски в колонке Number_of_rooms средним значением
df['Number_of_rooms'] = df['Number_of_rooms'].fillna(int(df['Number_of_rooms'].mean()))
```

```python
# Обрабатываем колонку area_of_rooms_m2
# Заменим чёрточки и плюсики в столбцах
df['area_of_rooms_m2'] = df['area_of_rooms_m2'].str.replace('-', ' ').str.replace('+', ' ').str.split(' ')
df['area_of_rooms_m2']
```

```python
# Напишем функцию котоая вернёт сумму площадей комнат и пропуск там где пропуск
def sum_areas(val):
    if val is None:
        return None  
    if isinstance(val, str):
        return float(val) if val.isdigit() else None  
    elif isinstance(val, list):
        return sum(float(x) for x in val if x.isdigit())  
    return None
```

```python
df['area_of_rooms_m2'] = df['area_of_rooms_m2'].apply(sum_areas)
```

```python
# Проверим на выбросы
df['area_of_rooms_m2'].max()
```

```python
# Избавимся от выбросов, на место выбросов вернём NaN

df.loc[(df.area_of_rooms_m2 < 10) | (df.area_of_rooms_m2 > 1000), 'area_of_rooms_m2'] = np.nan
```

```python
df['area_of_rooms_m2'].max()
```

```python
df['area_of_rooms_m2'] = df['area_of_rooms_m2'].fillna(int(df['area_of_rooms_m2'].mean()))
```

```python
# Обработаем колонку Ceiling height_m
# Проверим её на выбросы
df['Ceiling_height_m'].max()
```

```python
# Избавимся от выбросов

df.loc[(df.Ceiling_height_m < 1.5) | (df.Ceiling_height_m > 4), 'Ceiling_height_m'] = np.nan
```

```python
df['Ceiling_height_m'].max()
```

```python
# Заполним пропуски средним
df['Ceiling_height_m'] = df['Ceiling_height_m'].fillna(int(df['Ceiling_height_m'].mean()))
```

```python
# Заполним пропуски в категориальных признаках модой
df['Metro'] = df['Metro'].fillna(df['Metro'].mode()[0])
df['Parking'] = df['Parking'].fillna(df['Parking'].mode()[0])
df['Interior_finishing'] = df['Interior_finishing'].fillna(df['Interior_finishing'].mode()[0])
df['Balcony'] = df['Balcony'].fillna(df['Balcony'].mode()[0])
df['Windows'] = df['Windows'].fillna(df['Windows'].mode()[0])
df['Bathroom'] = df['Bathroom'].fillna(df['Bathroom'].mode()[0])
df['Children/pets_allowed'] = df['Children/pets_allowed'].fillna(df['Children/pets_allowed'].mode()[0])
df['Additionally'] = df['Additionally'].fillna(df['Additionally'].mode()[0])
df['Elevator'] = df['Elevator'].fillna(df['Elevator'].mode()[0])
df['Garbage_chute'] = df['Garbage_chute'].fillna(df['Garbage_chute'].mode()[0])
df['Name_of_the_residential_complex'] = df['Name_of_the_residential_complex'].fillna(df['Name_of_the_residential_complex'].mode()[0])
df['Home_series'] = df['Home_series'].fillna(df['Home_series'].mode()[0])
df['Link_to_ad'] = df['Link_to_ad'].fillna(df['Link_to_ad'].mode()[0])
```

# Feature engineering

```python
df = df.drop_duplicates(subset=['Address', 'Price', 'Additionally'], keep='last')
```

удалим колонку Type, т.к. тип всех объявлений - "Квартира"

```python
df = df.drop(['Type'], axis=1)
```

удалим колонки с уникальной идентификационной информацией для каждого объявления, кроме ID

```python
df = df.drop(['Address', 'Phone_number', 'Description'], axis=1)
```

преобразуем колонку Metro в две колонки: Location c названием станции и Walk_time с временем в пути до метро

```python
df[['Location', 'Walk_time']] = df['Metro'].str.extract(r'м\. (.*?) \((\d+) мин пешком\)')
```

```python
# удалим исходную колонку Metro
df = df.drop(['Metro'], axis=1)
```

```python
# заполним пропуски в колонках Location и Walk_time
df['Location'] =df['Location'].fillna(df['Location'].mode()[0])
df['Walk_time'] = pd.to_numeric(df['Walk_time'], errors='coerce')
df['Walk_time'] = df['Walk_time'].fillna(int(df['Walk_time'].mean()))
```

преобразуем категориальный признак - название метро -  в числовой:

```python
metro = {row: i for i, row in enumerate(sorted(list(set(df['Location']))))}
df['Location'] = [metro[row] for row in df['Location']]
```

```python
df[['Floor', 'Numb_of_floors', 'Building_type']] = df['Building_info'].str.extract(r'(\d+)/(\d+)(?:,\s*(.+))?')
```

```python
# избавимся от пропусков Building_type, удалим исходную колонку Building_info
df['Building_type'].fillna('Не указано', inplace=True)
df = df.drop(['Building_info'], axis=1)
```

преобразуем категориальный признак - тип дома -  в числовой:

```python
b_type = {row: i for i, row in enumerate(sorted(list(set(df['Building_type']))))}
df['Building_type'] = [b_type[row] for row in df['Building_type']]
```

преобразуем категориальные признаки - парковка и ремонт - в числовой:

```python
parking = {row: i for i, row in enumerate(sorted(list(set(df['Parking']))))}
df['Parking'] = [parking[row] for row in df['Parking']]
```

```python
interior = {row: i for i, row in enumerate(sorted(list(set(df['Interior_finishing']))))}
df['Interior_finishing'] = [interior[row] for row in df['Interior_finishing']]
```

разобьем колонку Price на несколько колонок:

```python
df['Pledge'] = df['Price'].str.extract(r'Залог - (\d+\.?\d*)\sруб\.')
df['Utilities'] = df['Price'].str.contains('Коммунальные услуги включены').astype(int)
df['Period'] = df['Price'].str.extract(r'Срок аренды - (.*?)(?:,|$)')
df['Prepayment'] = df['Price'].str.contains('Предоплата').astype(int)
df['Price'] = df['Price'].str.split(' ').str[0]
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

df['Pledge'].fillna(0, inplace=True)
df['Pledge'] = df['Pledge'].astype(float)
df['Utilities'] = df['Utilities'].astype(int)
df['Prepayment'] = df['Prepayment'].astype(int)
```

преобразуем колонки "Floor" и "Numb_of_floors" в числовой вид:

```python
df['Floor'] = df['Floor'].astype(int)
df['Numb_of_floors'] = df['Numb_of_floors'].astype(int)
```

преобразуем категориальный признак - период сдачи квартиры -  в числовой:

```python
period = {row: i for i, row in enumerate(sorted(list(set(df['Period']))))}
df['Period'] = [period[row] for row in df['Period']]
```

```python
# Обрабатываем колонку Balcony
df['Loggia'] = [row.split(' ')[-1] if (row.split(' ')[-2]) == 'Лоджия' else '(0)' for row in df['Balcony']]
df['Balcony'] = [row.split(' ')[1] if row.split(' ')[0] == 'Балкон' else '(0)' for row in df['Balcony']]
```

```python
# Корректируем значения, избавляясь от скобок и переделывая из str в int
df['Loggia'] = [row.split('(')[1] for row in df['Loggia']]
df['Loggia'] = [int(row.split(')')[0]) for row in df['Loggia']]
df['Balcony'] = [row.split('(')[1] for row in df['Balcony']]
df['Balcony'] = [int(row.split(')')[0]) for row in df['Balcony']]
```

```python
# Обрабатываем колонку Bathroom 	
df['Bathroom_combined'] = [row.split(' ')[-1] if (row.split(' ')[-2]) == 'Совмещенный' else '(0)' for row in df['Bathroom']]
df['Bathroom_separated'] = [row.split(' ')[1] if row.split(' ')[0] == 'Раздельный' else '(0)' for row in df['Bathroom']]
```

```python
# Корректируем значения, избавляясь от скобок и переделывая из str в int
df['Bathroom_combined'] = [row.split('(')[1] for row in df['Bathroom_combined']]
df['Bathroom_combined'] = [int(row.split(')')[0]) for row in df['Bathroom_combined']]
df['Bathroom_separated'] = [row.split('(')[1] for row in df['Bathroom_separated']]
df['Bathroom_separated'] = [int(row.split(')')[0]) for row in df['Bathroom_separated']]
```

```python
# Обработаем колонку Elevator
df['Elevator_pass'] = [row.split(' ')[-1] if (row.split(' ')[-2]) == 'Пасс' else '(0)' for row in df['Elevator']]
df['Elevator_cargo'] = [row.split(' ')[1] if row.split(' ')[0] == 'Груз' else '(0)' for row in df['Elevator']]
```

```python
# Корректируем значения, избавляясь от скобок и переделывая из str в int
df['Elevator_pass'] = [row.split('(')[1] for row in df['Elevator_pass']]
df['Elevator_pass'] = [int(row.split(')')[0]) for row in df['Elevator_pass']]
df['Elevator_cargo'] = [row.split('(')[1] for row in df['Elevator_cargo']]
df['Elevator_cargo'] = [int(row.split(')')[0]) for row in df['Elevator_cargo']]
```

```python
# Удалим лишние столбцы
df.drop(columns=['Bathroom', 'Elevator'], inplace=True)
```

```python
# Обработаем колонку Windows
df['Windows_street'] = [1 if (row.split(' ')[1]) == 'улицу' else 0 for row in df['Windows']]
df['Windows_yard'] = [1 if row.split(' ')[-1] == 'двор' else 0 for row in df['Windows']]
```

```python
# удалим колонку Windows
df.drop(columns=['Windows'], inplace=True)
```

```python
# Обработаем колонку Children/pets_allowed
df['Children_allowed'] = [1 if (row.split(' ')[2]) == 'детьми' else 0 for row in df['Children/pets_allowed']]
df['Pets_allowed'] = [1 if row.split(' ')[-1] == 'животными' else 0 for row in df['Children/pets_allowed']]
```

```python
# Удалим лишний столбец Children/pets_allowed
df.drop(columns=['Children/pets_allowed'], inplace=True)
```

```python
# Удалим столбец с ссылкой на объявление, так как он не несёт в себе никакой информации, а так же удалим столбцы Name_of_the_residential_complex и Home_series, так как в них была большая доля пропусков и по смыслу они не сильно значимы
df.drop(columns=['Name_of_the_residential_complex', 'Home_series', 'Link_to_ad'], inplace=True)
```

```python
# Отредактируем колонку Garbage_chute
df['Garbage_chute'] = [1 if row == 'Да' else 0 for row in df['Garbage_chute']]
```

```python
# Преобразуем колонку Additionally
df['Furniture_in_the_rooms'] = [1 if 'Мебель в комнатах' in row else 0 for row in df['Additionally']]
df['Kitchen_furniture'] = [1 if 'Мебель на кухне' in row else 0 for row in df['Additionally']]
df['Bath'] = [1 if 'Ванна' in row else 0 for row in df['Additionally']]
df['Shower_cabin'] = [1 if 'Душевая кабина' in row else 0 for row in df['Additionally']]
df['Washing_machine'] = [1 if 'Стиральная машина' in row else 0 for row in df['Additionally']]
df['Dishwasher'] = [1 if 'Посудомоечная машина' in row else 0 for row in df['Additionally']]
df['TV'] = [1 if 'Телевизор' in row else 0 for row in df['Additionally']]
df['Fridge'] = [1 if 'Холодильник' in row else 0 for row in df['Additionally']]
df['Internet'] = [1 if 'Интернет' in row else 0 for row in df['Additionally']]
df['Сonditioner'] = [1 if 'Кондиционер' in row else 0 for row in df['Additionally']]
df['Phone'] = [1 if 'Телефон' in row else 0 for row in df['Additionally']]
```

```python
# Избавимся от лишнего столбца
df.drop(columns=['Additionally'], inplace=True)
```
