```python
import numpy as np
import pandas as pd
```

# Исходный Data Frame

```python
path = './_data.csv'
data = pd.read_csv(path)
part_1 = data.iloc[:, :13]
part_1.head()
```

# Преобразование названий колонок

```python
part_1.columns = ['Serial_number', 'Announcement ID', 'Number_of_rooms', 'Type', 'Metro', 'Address', 'Area,m2', 'Building_info', 'Parking', 'Price', 'Phone_number', 'Description', 'Interior_finishing']
```

```python
part_1.head()
```

# Избавляемся от пропусков

```python
part_1_re = part_1
part_1_re['Price'] = part_1_re['Price'].str.split(' ').str[0]
part_1_re['Area,m2'] = part_1_re['Area,m2'].str.split('/').str[0]
part_1_re['Number_of_rooms'] = part_1_re['Number_of_rooms'].str.split(',').str[0]
part_1_re['Price'] = pd.to_numeric(part_1_re['Price'], errors='coerce')
part_1_re['Area,m2'] = pd.to_numeric(part_1_re['Area,m2'], errors='coerce')
part_1_re['Number_of_rooms'] = pd.to_numeric(part_1_re['Number_of_rooms'], errors='coerce')
part_1_re.head()
```

```python
# Заолним пропуски в колонке Number_of_rooms средним значением
part_1_re['Number_of_rooms'] = part_1_re['Number_of_rooms'].fillna(int(part_1_re['Number_of_rooms'].mean()))
```

```python
# Заполним пропуски в категориальных признаках модой
part_1_re['Metro'] = part_1_re['Metro'].fillna(part_1_re['Metro'].mode()[0])
part_1_re['Parking'] = part_1_re['Parking'].fillna(part_1_re['Parking'].mode()[0])
part_1_re['Interior_finishing'] = part_1_re['Interior_finishing'].fillna(part_1_re['Interior_finishing'].mode()[0])
```
